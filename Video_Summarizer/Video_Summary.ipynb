```python
# Youtube link: https://www.youtube.com/watch?v=yNnvWem8V3U
import moviepy.editor as mp

myclip = mp.VideoFileClip(r"Audio_Files\Video1.mp4")
myclip.audio.write_audiofile(r"Audio_Files\Audio1.mp3")
```

```python
import requests

filename = 'Audio_Files\Audio1.mp3'
api_key =  '702215a26cfa4411a4edf2238358b0c6'
upload_endpoint = 'https://api.assemblyai.com/v2/upload'

def read_file(filename, chunk_size=5242880):
    with open(filename, 'rb') as f:
        while True:
            data = f.read(chunk_size)
            if not data:
                break
            yield data


headers = {'authorization': api_key,
           'content-type': 'application/json'}
response = requests.post(upload_endpoint,
                         headers=headers,
                         data=read_file(filename))

audio_url = response.json()['upload_url']
```

```python
import pprint

TRANSCRIPT_ENDPOINT = 'https://api.assemblyai.com/v2/transcript'

response = requests.post(TRANSCRIPT_ENDPOINT, headers={'authorization':api_key,'content-type':'application/json'},
                         json={'audio_url':audio_url,'auto_chapters':True},)

response_json = response.json()
print(pprint.pprint(response_json))
```

```python
TRANSCRIPT_ENDPOINT = 'https://api.assemblyai.com/v2/transcript/'+response_json['id']

api_key =  '702215a26cfa4411a4edf2238358b0c6'
response = requests.get(TRANSCRIPT_ENDPOINT, headers={'authorization':api_key},)

response_json = response.json()

print(pprint.pprint(response_json))
```

```python
chapters = response_json['chapters']

fobj = open('Audio_Files\Summary1.txt','w')
var=''

for i in range(len(chapters)):
    var += chapters[i]['summary']+'\n'

fobj.write(var)
fobj.close()
```
